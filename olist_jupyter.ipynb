```python
# SQL QUERY
master_query = """
WITH OrderAggregates AS (
    SELECT 
        c.customer_unique_id,
        o.order_id,
        o.order_purchase_timestamp,
        SUM(p.payment_value) AS total_order_value
    FROM orders o
    JOIN customers c 
        ON o.customer_id = c.customer_id
    JOIN order_payments p 
        ON o.order_id = p.order_id
    WHERE o.order_status = 'delivered'
    GROUP BY 
        c.customer_unique_id, 
        o.order_id, 
        o.order_purchase_timestamp
)
SELECT 
    customer_unique_id,
    order_id,
    order_purchase_timestamp,
    total_order_value,
    ROW_NUMBER() OVER(
        PARTITION BY customer_unique_id 
        ORDER BY order_purchase_timestamp
    ) AS customer_order_rank
FROM OrderAggregates;
"""

# Execute the query and load it into a DataFrame called rfm_raw
print("Extracting master transaction table from MySQL...")
rfm_raw = pd.read_sql(master_query, engine)

# Check the results
print(f"Extraction complete! Pulled {len(rfm_raw)} completed orders.")
rfm_raw.head()
```

```python
import datetime as dt

# Converting the string/object type into a proper Pandas datetime type.
rfm_raw['order_purchase_timestamp'] = pd.to_datetime(rfm_raw['order_purchase_timestamp'])

# Setting a Reference Date 
# We use the most recent purchase in the dataset + 1 day(cause the recent purchaser's recency becomes 0 without +1)
reference_date = rfm_raw['order_purchase_timestamp'].max() + dt.timedelta(days=1)

# Calculation
rfm_calc = rfm_raw.groupby('customer_unique_id').agg({
    
    # Recency: Days between "today" and their last purchase
    'order_purchase_timestamp': lambda x: (reference_date - x.max()).days,
    
    # Frequency: The highest 'customer_order_rank'
    'customer_order_rank': 'max',
    
    # Monetary: Sum of all their orders
    'total_order_value': 'sum'
}).reset_index()

# Rename the columns so they are clean and readable
rfm_calc.rename(columns={
    'order_purchase_timestamp': 'Recency',
    'customer_order_rank': 'Frequency',
    'total_order_value': 'Monetary'
}, inplace=True)
# inplace=True is --> "Modify the existing DataFrame directly instead of creating and returning a new DataFrame."
```

```python
# Recency Score: Lower days is BETTER (5 is the most recent)
rfm_calc['R_Score'] = pd.qcut(rfm_calc['Recency'], q=5, labels=[5, 4, 3, 2, 1])

# Frequency Score: Higher count is BETTER (5 bought the most often)
rfm_calc['F_Score'] = pd.qcut(rfm_calc['Frequency'].rank(method='first'), q=5, labels=[1, 2, 3, 4, 5])

# Monetary Score: Higher spend is BETTER (5 spent the most money)
rfm_calc['M_Score'] = pd.qcut(rfm_calc['Monetary'], q=5, labels=[1, 2, 3, 4, 5])

# Combine the three scores into a single RFM String (e.g., '555' or '124')
rfm_calc['RFM_Segment_Code'] = rfm_calc['R_Score'].astype(str) + rfm_calc['F_Score'].astype(str) + rfm_calc['M_Score'].astype(str)

rfm_calc.head()
```

```python
def assign_segment(row):
    # Converting string scores back to integers for logical comparisons
    r = int(row['R_Score'])
    f = int(row['F_Score'])
    m = int(row['M_Score'])

    # Champions: Bought recently, buy often, and spend a lot
    if r >= 4 and f >= 4 and m >= 4:
        return 'Champions'
    
    # Loyal Customers: Consistent buyers
    elif r >= 3 and f >= 3:
        return 'Loyal Customers'
    
    # At Risk: Big spenders and frequent buyers, but haven't purchased in a long time
    elif r <= 2 and f >= 3 and m >= 4:
        return 'At Risk'
    
    # Recent Customers: Bought very recently, but haven't bought much or often yet
    elif r >= 4 and f <= 2:
        return 'New Customers'
    
    # Hibernating: Last purchase was long ago, low spend, low frequency
    elif r <= 2 and f <= 2:
        return 'Lost'
    
    # Catch-all for middle-of-the-pack customers
    else:
        return 'Average/Regular'

# Applying the function to create our final labels
rfm_calc['Customer_Segment'] = rfm_calc.apply(assign_segment, axis=1)

# Printing the final count of customers in each segment
print(rfm_calc['Customer_Segment'].value_counts())

# rfm_calc.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as ticker

# Set a clean, professional style
sns.set_theme(style="whitegrid")

# 1. Prepare the data
segment_counts = rfm_calc['Customer_Segment'].value_counts().reset_index()
segment_counts.columns = ['Customer_Segment', 'Count']

# 2. Build the plot (UPDATED WITH HUE)
plt.figure(figsize=(10, 6))
ax = sns.barplot(
    data=segment_counts, 
    y='Customer_Segment', 
    x='Count', 
    hue='Customer_Segment', # <--- The Fix
    palette='viridis', 
    legend=False            # <--- The Fix
)

plt.title('Customer Base Distribution by Segment', fontsize=16, fontweight='bold', pad=15)
plt.xlabel('Number of Customers', fontsize=12)
plt.ylabel('')

# Add the exact numbers to the end of the bars
for p in ax.patches:
    ax.annotate(f'{int(p.get_width()):,}', 
                (p.get_width(), p.get_y() + p.get_height() / 2.), 
                ha='left', va='center', 
                xytext=(5, 0), textcoords='offset points',
                fontsize=10)

plt.xlim(0, segment_counts['Count'].max() * 1.15)
plt.show()
```

```python
# 1. Prepare the data
segment_revenue = rfm_calc.groupby('Customer_Segment')['Monetary'].sum().reset_index()
segment_revenue = segment_revenue.sort_values(by='Monetary', ascending=False)

# 2. Build the plot (UPDATED WITH HUE)
plt.figure(figsize=(10, 6))
ax = sns.barplot(
    data=segment_revenue, 
    y='Customer_Segment', 
    x='Monetary', 
    hue='Customer_Segment', # <--- The Fix
    palette='magma',
    legend=False            # <--- The Fix
)

plt.title('Total Revenue Driven by Customer Segment', fontsize=16, fontweight='bold', pad=15)
plt.xlabel('Total Revenue ($)', fontsize=12)
plt.ylabel('')

# Format the x-axis to show clean currency (Millions)
formatter = ticker.FuncFormatter(lambda x, pos: f'${x/1e6:.1f}M')
ax.xaxis.set_major_formatter(formatter)

plt.show()
```

```python
# 1. Create a pivot table 
# Rows = Recency Score, Columns = Frequency Score, Values = Average Spend
rfm_pivot = rfm_calc.pivot_table(
    index='R_Score', 
    columns='F_Score', 
    values='Monetary', 
    aggfunc='mean'
)

# Reverse the index so the best Recency Score (5) is at the top of the chart
rfm_pivot = rfm_pivot.sort_index(ascending=False)

# 2. Build the plot
plt.figure(figsize=(8, 6))
sns.heatmap(
    rfm_pivot, 
    annot=True,          # Put the numbers in the squares
    fmt=".0f",           # Remove decimals
    cmap="YlGnBu",       # Yellow-Green-Blue color map
    cbar_kws={'label': 'Average Customer Spend ($)'}
)

plt.title('Average Spend Matrix: Recency vs. Frequency', fontsize=16, fontweight='bold', pad=15)
plt.xlabel('Frequency Score (Higher = Buys More Often)', fontsize=12)
plt.ylabel('Recency Score (Higher = Bought More Recently)', fontsize=12)

plt.show()
```
